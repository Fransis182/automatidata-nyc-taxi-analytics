# 06 — Final Recommendations

This notebook serves as the final wrap-up of the Automatidata project.

## Executive conclusion

This project showed that the NYC taxi data can support several valuable analytical tasks once quality issues are identified and managed carefully.

### What worked well
- the dataset was structurally complete
- key trip variables supported strong modeling performance
- the regression model provided useful fare estimates
- the classification model delivered balanced predictive power

### Main risks
- outliers and invalid values can distort insights
- cash tipping behavior is not fully observed
- classification outputs can be misused if treated as deterministic

### Recommendation
Use the fare model as an operational estimation baseline and treat the tipping classifier as a decision-support signal only, never as a tool to deny or reduce service quality.


## Suggested stakeholder message

- Continue validating data quality before retraining
- Tune thresholds according to the cost of false positives vs false negatives
- Validate by segment, especially location and time
- Keep ethical safeguards explicit in any deployment discussion
