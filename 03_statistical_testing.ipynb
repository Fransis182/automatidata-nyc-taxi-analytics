# 03 — Statistical Testing

This notebook evaluates whether `fare_amount` differs significantly between card and cash payments.

```
import pandas as pd
from scipy import stats
```

```
df = pd.read_csv("../data/2017_Yellow_Taxi_Trip_Data.csv")
df.head()
```

```
card = df[df["payment_type"] == 1]["fare_amount"]
cash = df[df["payment_type"] == 2]["fare_amount"]

card_mean = card.mean()
cash_mean = cash.mean()
difference = card_mean - cash_mean

t_stat, p_value = stats.ttest_ind(card, cash, equal_var=False)

card_mean, cash_mean, difference, t_stat, p_value
```

## Validated result from the completed workflow

- Mean fare (card): **~13.43**
- Mean fare (cash): **~12.21**
- Difference: **~1.22**
- t-statistic: **~6.87**
- p-value: **~6.8e-12**

Interpretation: the difference in average `fare_amount` between card and cash payments is statistically significant at α = 0.05.
