## Data quality checks

```
# Highest and lowest trip distances
df.sort_values(by="trip_distance", ascending=False).head(10), df.sort_values(by="trip_distance", ascending=True).head(10)
```

```
# Highest and lowest total amounts
df.sort_values(by="total_amount", ascending=False).head(10), df.sort_values(by="total_amount", ascending=True).head(10)
```

```
# Payment type counts
df["payment_type"].value_counts().sort_index()
```

```
# Mean tip by payment type
df.groupby("payment_type")["tip_amount"].mean()
```

```
# Vendor counts and average total amount
df.groupby("VendorID").agg(
    count=("VendorID", "size"),
    mean_total_amount=("total_amount", "mean")
)
```

```
# Tip amount by passenger_count for card payments
card_df = df[df["payment_type"] == 1]
card_df.groupby("passenger_count")["tip_amount"].mean().sort_index()
```
